```python
import pandas as pd
from matplotlib import pyplot as plt
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]
plt.plot(x, y)
plt.plot(x, z)
plt.title("Problema 1.1")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["este es y", "este es z"])
plt.show()
```

```python
# graficando countries
```

```python
country = pd.read_csv("countries.csv")
```

```python
country
```

```python
Honduras = country[country.country == "Honduras"]
```

```python
Ecuador = country[country.country == "Ecuador"]
```

```python
plt.plot(Honduras.year, Honduras.population/10**6)
plt.plot(Ecuador.year, Ecuador.population/10**6)
plt.legend(["Ecuador", "Honduras"])
plt.title("Poblacion Honduras y Ecuador")
plt.xlabel("Año")
plt.ylabel("Millones de Habitantes")
plt.show()
```
